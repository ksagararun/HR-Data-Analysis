#HR DATA ANALYSIS

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
Data = pd.read_csv("HR Data.csv")
```

Exploring Data

```python
# Display top 5 rows dataset
Data.head()
```

```python
# Display last 5 rows of dataset
Data.tail()
```

```python
# Display basic information about the dataset
Data.info()
```

```python
# Display summary statistics
Data.describe()
```

```python
# Check for missing values
Data.isnull().sum()
```

```python
# Columns to remove
columns_to_remove = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']

# Drop unnecessary columns
Data = Data.drop(columns=columns_to_remove)
```

```python
#overview of Data after removing unwanted columns
Data.info()
```

```python
# Eliminate redundant entries
Data = Data.drop_duplicates()

# Display the modified Data without duplicates
print(Data)
```

sanitizing specific columns.

```python
# Map 'Male' and 'Female' to 0 and 1 in the 'Gender' column
Data['Gender'] = Data['Gender'].map({'Male': 0, 'Female': 1})
```

```python
# Sanitize 'Age' column by clipping values between 18 and 60
Data['Age'] = Data['Age'].clip(lower=18, upper=60)
```

Visualize Data

```python
# Visualizing Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(Data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing Department distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Department', data=Data, palette='pastel')
plt.title('Distribution of Departments')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=Data, palette='muted')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

#NOTE :- 0 : Male, 1 : Female
```

```python
# Visualizing Attrition distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition', data=Data, palette='Set2')
plt.title('Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```
